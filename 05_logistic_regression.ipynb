## Machine learning

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
df_short = df[:10000]
X= df_short.loc[:, df.columns != 'pm']
y= round(df_short['pm'])

from sklearn import preprocessing
from sklearn import utils
```

It is very important to mention that I worked only with 10k rows, because the dataset is huge.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=2)
arr = []
for i in [10,100,1000,5000,10000,15000,20000]:
    print(f"got {i}")
    regr = LogisticRegression(max_iter = i)
    regr.fit(X_train, y_train)
    arr.append(regr.score(X_test, y_test))
```

```python
print(arr)

fig = plt.figure(figsize =(10, 7))
 
plt.plot([1,2,3,4,5,6,7],arr)
plt.ylabel("Accuracy")
plt.xlabel("Number of iters ([10,100,1000,5000,10000,15000,20000])")
plt.title("Accuracy throught iters in ml model")

plt.show()
```

As seen, in the first case, when reaching to 20000 iterations, accuracy remains the same, thus it means that the convergence stops, and the highest accuracy achieved is "0.59"

## Trying the correlating values with the target column (pm)

```python
DF = df[["stator_tooth",    "stator_winding",  "stator_yoke",       "ambient",           "coolant","pm"]]
DF
```

```python
df_crop_1 = DF[:10000]
X= df_crop_1[["stator_tooth",    "stator_winding",  "stator_yoke",       "ambient",           "coolant"]]
y= round(df_crop_1['pm'])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=2)
arr = []
for i in [10,100,1000,5000,10000,15000,20000,30000]:
    print(f"working on {i} iters")
    regr = LogisticRegression(max_iter = i)
    regr.fit(X_train, y_train)
    arr.append(regr.score(X_test, y_test))
```

```python
print(arr)
```

```python
fig = plt.figure(figsize =(10, 7))
 
plt.plot([1,2,3,4,5,6,7,8],arr)
plt.ylabel("Accuracy")
plt.xlabel("Number of iters ([10,100,1000,5000,10000,15000,20000])")
plt.title("Accuracy throught iters in ml model")

plt.show()
```

Here we see that the accuracy is with one percent bigger than usual.


```python
df_crop_1 = DF[:5000]
X= df_crop_1[["stator_tooth",    "stator_winding",  "stator_yoke",       "ambient",           "coolant"]]
y= round(df_crop_1['pm'])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=2)

regr = LogisticRegression(max_iter = 10000)
regr.fit(X_train, y_train)
print(regr.score(X_test, y_test))
```

However, here we can see that if we make the dataset that we are working on, smaller, the accuracy is getting much bigger. Thus, we can deduce that the optimal number of rows to train the model is above 5000, on 10000 or more.

## Conclusion
As seen, when working with all the data in the dataset, we get the maximal accuracy of 0.59. Though, when working with most correlative data, we can get 0.60 or even 0.7115 accuracy for the ML model.
This means that the accuracy of the model, depends not only on how good the data is preprocessed, but in number of rows that our model is trained on, and also number of iterations (you can see those two graphs, an the correlation between num. of iters and accuracy)
