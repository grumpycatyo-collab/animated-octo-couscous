```python
import pandas as pd
df = pd.read_csv("heart.csv")
df
```

```python
mean_value = df['trestbps'].mean()
print("Mean:", mean_value)
median_value = df['trestbps'].median()
print("Median:", median_value)
mode_value = df['trestbps'].mode()
print("Mode:", mode_value)
std_value = df['trestbps'].std()
print("Standard Deviation:", std_value)
var_value = df['trestbps'].var()
print("Variation:", var_value)
range_value = df['trestbps'].max() - df['trestbps'].min()
print("Range of Values:", range_value)
```

```python
corr_matrix = df.corr()
print(corr_matrix["target"].sort_values(ascending=False))
```
